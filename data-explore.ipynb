```python
data = pd.read_csv('data/cleaned-reviews.csv')
```

```python
data.groupby('rating_review').count()
```

```python
data = data[data['rating_review'] != 3]
data
```

```python
def label_examples(rating):
    return 0 if rating < 3 else 1
```

```python
data['good_rating'] = data['rating_review'].apply(label_examples)
```

```python
data.drop('rating_review', axis = 1, inplace = True)
```

```python
data.groupby('good_rating').count()
```

```python
data.to_csv('data/binary-cleaned-data.csv', index = False)
```
